```python
import altair as alt
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/building_inventory.csv", 
na_values = {"Year Acquired":0,
            "Year Constructed":0,
            "Square Footage":0,
            "Total Floors":0
                })
```

```python
df.columns
```

```python
df = df[df['Year Constructed']>2000]
```

```python
df
```

```python
alt.data_transformers.disable_max_rows()
```

```python
click = alt.selection_point(encodings=['color'])
```

```python
scatter = alt.Chart(df).mark_point().encode(
    x=alt.X('Year Constructed:Q', scale=alt.Scale(domain=(2000, 2020))),
    y='Square Footage:Q',
    color=alt.condition(click,
                        alt.Color('Usage Description:N', scale=alt.Scale(scheme='tableau20')),
                        alt.value('lightgray')),
    tooltip=['Year Constructed', 'Square Footage', 'Usage Description']
).add_params(
    click
)
```

```python
bars = alt.Chart(df).mark_bar().encode(
    x=alt.X('Year Constructed:O', axis=alt.Axis(labelAngle=45)),
    y='mean(Total Floors):Q',
    color=alt.condition(click, alt.value("steelblue"), alt.value("lightgray"))
).transform_filter(
    click
)
```

```python
chart = alt.vconcat(scatter, bars)
chart
```

```python
open("assignment.html", "w").write(alt.vconcat(scatter,bars).to_html())
```
